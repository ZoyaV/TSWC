```python

with open('datasets/original_db1564999251377360.pickle', 'rb') as f:
     data = np.asarray(pickle.load(f))
with open('datasets/st_labels_db1564999251377360.pickle', 'rb') as f:
     labels = pickle.load(f)
colors = ['r', 'g', 'b', 'y']
colors_plot = [colors[c] for c in labels]

data = data[:64*40]
```

```python
plt.figure(figsize = (24,4))
plt.plot(data)
```

### I-KMeans

```python
start_lvl = 9
deep = 2

ikmeans = IKMeans(data, start_lvl = start_lvl, k = 4)
plt.figure(figsize = (10,10))

for i in range(0, deep):
    plt.subplot(10,1,i+1)      
    plt.plot(ikmeans.cur_cA)
    
    ikmeans.next_lvl()
    ikmeans.fit()
```

```python
ikmeans.plot_clusters()
```

```python
patterns = ikmeans.clasters.sorted_patterns
```
